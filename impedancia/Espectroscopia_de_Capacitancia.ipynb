<h2>1. Importa las librerias</h2>

```
import numpy as np
import matplotlib.pyplot as plt

f = np.logspace(-3, 6, 1000)
w = 2 * np.pi * f
```

# 1. Modelo circuital impedimétrico

```
Cd = 100e-6 # F
Rd = 100 # Ohm
Rr = 1 / 7.748091729e-5 # Ohm -> 1/R = 2 q^2 / h
Rm = 100 # Ohm
w = 2 * np.pi * f 
Z_Cd = -1j/(w * Cd)

Zd = Rd + Z_Cd
Z1 = 1 / (1/Zd + 1/Rr)
Z = Rm + Z1

C = Z / (1j*w)
```

```
plt.figure(dpi=200)
plt.plot(Z.real, -Z.imag)
plt.xlabel('real{Z}')
plt.ylabel('-imag{Z}')
plt.title('Diagrama de Nyquist')
```

```
plt.figure(dpi=200)
plt.plot(C.real, -C.imag)
plt.xlabel('real{C}')
plt.ylabel('-imag{C}')
plt.title('Diagrama de Nyquist')
```

# 2. Modelo de Cole-Cole para elemento capacitivo

```
Co  = 120e-6 # F
Ci  =  15e-6 # F
tau =  95e-3 # s/rad
c = 0.5

C = Ci + (Co - Ci) / ( 1+ (1j*w*tau)**c )
Z = 1 / (1j*w*C)
```

```
plt.figure(dpi=200)
plt.plot(Z.real, -Z.imag)
plt.xlabel('real{Z}')
plt.ylabel('-imag{Z}')
plt.title('Diagrama de Nyquist')
```

```
plt.figure(dpi=200)
plt.plot(C.real, -C.imag)
plt.xlabel('real{C}')
plt.ylabel('-imag{C}')
plt.title('Diagrama de Nyquist')
```
